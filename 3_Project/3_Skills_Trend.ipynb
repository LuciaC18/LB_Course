## How are in-demand skills trending for Data Analysts?

### Methodogy

1. Aggregate skill counts monthly
2. Analyse based on percentage of total jobs
3. Plot the monthly skill demand

```python
# Importing libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Only focusing on Data Analyst roles in the US

df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
# Creating a new column to store posting month number
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python

df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(
    index='job_posted_month_no',
    columns='job_skills',
    aggfunc='size',
    fill_value=0)

df_DA_US_pivot
```

```python
# To be able to sort out the total count of skills, create a new row, called 'Total'

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()
df_DA_US_pivot
```

```python
# Sort the df by the total count of mentions.

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_US_pivot = df_DA_US_pivot.drop('Total')
df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)
df_DA_US_percent
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns=['job_posted_month_no', 'level_0', 'index'])

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False)
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


offsets = {
    'tableau': 1.5,
    'python': -1.5
}

for i in range(5):
    label = df_plot.columns[i]
    y_pos = df_plot.iloc[-1, i] + offsets.get(label, 0)
    plt.text(11.2, y_pos, label)
```
