## How well do jobs and skills pay for Data Analysts?


### Methodology

1. Evaluate median salary for top 6 data jobs
2. Find median salary per skill for Data Analysts
3. Visualize for highest paying skills and most demanded skills

```python
# Importing libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Filtering for the job title and country

df_US = df[(df['job_country'] == 'United States')].dropna(subset=['salary_year_avg'])
```

```python
# Getting the top 6 job titles (by number of job postings) and turning into a list use as a filter later on.

job_titles = df_US['job_title_short'].value_counts().index[:6].to_list()

job_titles
```

```python
df_US_top_6 = df_US[df_US['job_title'].isin(job_titles)]
df_US_top_6
```

```python
# Grouping by job titles, and then ordering by median salaries. This will allow the y axis in the plot to be sorted by salaries

job_order = df_US_top_6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
job_order
```

```python
sns.boxplot(data=df_US_top_6, x='salary_year_avg', y='job_title_short', order= job_order)
sns.set_theme(style='ticks')


plt.title('Salary Distributions in the United States')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 600000)
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()
```

## Investigate Median Salary v. Skill for Data Analysts

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
# Dropping any rows with no salary info

df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])
```

```python
# 'Exploding' skills lists

df_DA_US = df_DA_US.explode('job_skills')
df_DA_US[['salary_year_avg', 'job_skills']].head(5)
```

```python
# Find top paying skills

df_DA_top_pay  = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)

df_DA_top_pay = df_DA_top_pay.head(10)

df_DA_top_pay
```

```python
# Find most common skills in DA roles

df_DA_skills = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

df_DA_skills = df_DA_skills.head(10).sort_values(by='median', ascending=False)

df_DA_skills
```

```python
# Create two plots
fig, ax = plt.subplots(2, 1)

sns.set_theme(style='ticks')

# Top 10 Highest Paid Skills for Data Analysts
sns.barplot(data=df_DA_top_pay, x='median', y=df_DA_top_pay.index, ax=ax[0], palette='flare_r')
ax[0].legend().remove()

ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))


sns.barplot(data=df_DA_skills, x='median', y=df_DA_skills.index, ax=ax[1], palette='flare_r')
ax[1].legend().remove()

ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysts')
ax[1].set_ylabel('')
ax[1].set_xlabel('')
ax[1].set_xlabel('Median Salary (USD)')
ax[1].set_xlim(ax[0].get_xlim())
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

fig.tight_layout()
```
